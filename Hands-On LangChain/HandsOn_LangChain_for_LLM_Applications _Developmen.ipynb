```python
from langchain.embeddings import HuggingFaceEmbeddings

embedding = HuggingFaceEmbeddings(model_name="thenlper/gte-large")
```

```python
sentence1 = "My dog Rover likes to chase squirrels"
sentence2 = "Fluffy my cat, refuses to eat from a can"
sentence3 = "The chevy bolt accelerates to 60 mph in 6.7 seconds"
```

```python
embedding1 = embedding.embed_query(sentence1)
embedding2 = embedding.embed_query(sentence2)
embedding3 = embedding.embed_query(sentence3)
```

```python
from langchain.document_loaders import PyPDFLoader

# Load PDF
loaders = [
    # Duplicate documents on purpose - messy data
    PyPDFLoader(f"{youssef19_machine_learning_lectures_path}/MachineLearning-Lecture01.pdf"),
    PyPDFLoader(f"{youssef19_machine_learning_lectures_path}/MachineLearning-Lecture01.pdf"),
    PyPDFLoader(f"{youssef19_machine_learning_lectures_path}/MachineLearning-Lecture02.pdf"),
    PyPDFLoader(f"{youssef19_machine_learning_lectures_path}/MachineLearning-Lecture03.pdf")
]
docs = []
for loader in loaders:
    docs.extend(loader.load())
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter


text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 1500,
    chunk_overlap = 150
)


splits  = text_splitter.split_documents(docs)
```

```python
len(splits)
```

```python

from langchain.vectorstores import Chroma

persist_directory = r'docs/chroma/'

vector_db = Chroma.from_documents(
    documents=splits,
    embedding=embedding,
    persist_directory=persist_directory
)
```

```python
question = "What are the names and emails of the course TA?"

docs=vector_db.similarity_search(question,k=3)
```

### Duplicate Entries in PDFs

Let's take a look at the first two results — we can see that they’re **identical**.  
The reason we got two identical answers is that we **intentionally loaded a duplicate entry** from the PDFs.  

This is problematic because we now have the **same information in two different chunks**, and both chunks will be passed to the language model later on.  
There’s **no added value** in repeating the same content, and it would be **much better** to have a **distinct chunk** instead — allowing the model to learn from **unique information**.

### Failure Mode: Incorrect Source Retrieval

Another type of failure mode can also happen when we specify a certain source to answer our questions.  
Let’s take a look at the question:  
> *"What did they say about supervised learning in the second lecture?"*

When we get the docs for this, intuitively we would expect them to all be part of the **second lecture**.  
We can check this because we have information in the **metadata** about what lectures they came from.  
So, let’s loop over all the documents and print out the metadata.

---

```python
question = "what did they say about supervised machine learning in the third lecture?"
docs = vectordb.similarity_search(question,k=5)
for doc in docs:
    print(doc.metadata)
```

As you can see, there’s actually a **combination of results** — some from the **third lecture**, some from the **second lecture**, and some from the **first**.  

The intuition about why this is failing is that we want documents from **only the second lecture** as a piece of structured information,  
but we’re just doing a **semantic lookup based on embeddings**, where it creates an embedding for the whole sentence,  
and it’s probably a bit more focused on *supervised learning*.  

Therefore, we’re getting results that are probably pretty relevant to **regression**,  
and if we take a look at the **fifth document**, the one that comes from the **first lecture**,  
we can see that it **does mention supervised learning**.
