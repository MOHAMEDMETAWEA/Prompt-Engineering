# Loading YouTube Videos into Text

```python
from langchain_community.document_loaders.generic import GenericLoader
from langchain_community.document_loaders.parsers import OpenAIWhisperParser
from langchain_community.document_loaders.blob_loaders.youtube_audio import YoutubeAudioLoader
```

```python
!pip install langchain-openai -q yt-dlp groq langchain-core python-dotenv

import os
from groq import Groq
from langchain_core.documents import Document


url = "https://www.youtube.com/watch?v=5p248yoa3oE"
save_dir = "./youtube/"
os.makedirs(save_dir, exist_ok=True)

audio_path = os.path.join(save_dir, "audio.mp3")

!yt-dlp -x --audio-format mp3 -o "{audio_path}" "{url}"

with open(audio_path, "rb") as f:
    transcript = client.audio.transcriptions.create(
        model="whisper-1",
        file=f
    )

docs = [Document(page_content=transcript.text, metadata={"source": url})]

print(" Transcript Preview:\n")
print(docs[0].page_content[:500])
```
