```python
from langchain.chat_models import ChatOpenAI
from langchain.schema import HumanMessage

chat = ChatOpenAI(model_name="gpt-4o-mini", temperature=0.5)
```

```python
prompt = """
Suggest 2 ways to lose my weight.
""".strip()

response = chat([HumanMessage(content=prompt)])
print(response.content)
```

# HUGGING FACE


```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch

from langchain.llms import HuggingFacePipeline

model_id = "h2oai/h2ogpt-gm-oasst1-en-2048-open-llama-3b"

tokenizer = AutoTokenizer.from_pretrained(model_id, use_fast=False)

base_model = AutoModelForCausalLM.from_pretrained(model_id,
                                            load_in_4bit=True,
                                            device_map='auto',)
```

```python
pipe=pipeline("text-generation",
              model=base_model,
              tokenizer=tokenizer,
              max_length=200)
llms=HugginFacePipeline(pipeline=pipe)
```

```python
template = """Suggest 2 ways to lose weight.
Answer in **1 line per way only**, no extra details.
"""

print(llms(template))
```
