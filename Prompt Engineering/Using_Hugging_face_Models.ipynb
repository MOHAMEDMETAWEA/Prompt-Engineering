# HuggingFace Models 

```python
# Use a pipeline as a high-level helper

from transformers import pipeline
```

```python
# Load model directly

from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("MaziyarPanahi/calme-3.2-instruct-78b")
model = AutoModelForCausalLM.from_pretrained("MaziyarPanahi/calme-3.2-instruct-78b")
```

```python
messages = [
    {"role": "user", "content": "Who are you?"},
]

pipe = pipeline("text-generation",
                 model="MaziyarPanahi/calme-3.2-instruct-78b" ,
                    tokenizer="MaziyarPanahi/calme-3.2-instruct-78b",max_length=512)
pipe(messages)
```

```python
inputs = tokenizer("Write a python function to add two numbers.", return_tensors="pt")
outputs = model.generate(**inputs, max_new_tokens=100)
```
