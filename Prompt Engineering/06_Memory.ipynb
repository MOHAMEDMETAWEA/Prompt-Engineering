## OpenAI

```python
from dotenv import dotenv_values
env_values = dotenv_values("./app.env")

openai_api_key = env_values['OPEN_API_KEY']
```

```python
from langchain.llms import OpenAI

llm = OpenAI(openai_api_key=openai_api_key,
             model_name="text-davinci-003",
             temperature=0.5)
```

## Buffer Memory

```python
from langchain.chains import ConversationChain
from langchain.callbacks import get_openai_callback
from langchain.memory import ConversationBufferMemory

conversation = ConversationChain(
    llm=llm,
    memory=ConversationBufferMemory()
)
```

```python
message_1 = "Can you list 3 places in Paris to visit?"

message_2 = "Which place has a lake view?"

message_3 = "Can I visit in winter?"
```

```python
output_1 = conversation.predict(input=message_1)
print(output_1)
```

```python
output_2 = conversation.predict(input=message_2)
print(output_2)
```

```python
for msg in conversation.memory.chat_memory.messages:
    print(msg)
    print('---------------')
```

```python
conversation.memory.clear()
```

```python
with get_openai_callback() as cb:
    print( conversation.predict(input=message_1) )

    print("======================")

    print(cb)
```

```python
conversation.memory.clear()
```

```python
conversation.verbose = True
```

```python
conversation.predict(input=message_1)
```

```python
conversation.predict(input=message_2)
```

```python
with get_openai_callback() as cb:
    print( conversation.predict(input=message_3) )

    print("======================")

    print(cb)
```

## Save to File

```python
from langchain.memory import ConversationSummaryMemory
from langchain.schema import messages_from_dict, messages_to_dict
import json

conversation = ConversationChain(
    llm=llm,
    memory=ConversationSummaryMemory(llm=llm),
    verbose=True
)
```

```python
message_1 = "Can you list 3 places in Paris to visit?"
message_2 = "Which place has a lake view?"

output_1 = conversation.predict(input=message_1)
output_2 = conversation.predict(input=message_2)
```

```python
# Not Supported
# conversation.save("/content/conversation-memory.json")
```

```python
# save messages to a file

dicts = messages_to_dict( conversation.memory.chat_memory.messages )

with open("/content/conversation-memory.json", "w") as dest:
    dest.write(json.dumps( dicts, ensure_ascii=False ))
```

```python
# load messages from a file
from langchain.memory import ChatMessageHistory

with open("/content/conversation-memory.json") as src:
    saved_history = json.loads(src.read())

history = ChatMessageHistory()
history.messages = messages_from_dict(saved_history)
```

```python
memory = ConversationSummaryMemory(llm=llm)
memory = memory.from_messages(chat_memory=history, llm=llm)

conversation = ConversationChain(
    llm=llm,
    memory=memory,
    verbose=True
)
```

```python
message_3 = "Can I visit in winter?"

print( conversation.predict(input=message_3) )
```
