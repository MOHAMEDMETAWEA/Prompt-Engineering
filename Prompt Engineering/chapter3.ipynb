# Prompt Templates

## What is a Prompt Template?
- A **predefined prompt structure** containing variables that can change with each LLM call.
- Purpose: **reuse the same structure dynamically**, e.g., changing topic or language.

## Types / Variations of Templates
- **Single-variable template**: contains only one variable (e.g., `topic`).  
- **Multi-variable template**: contains multiple variables (e.g., `topic` and `language`).  
- **Multi-line templates**: text can be split into multiple lines to organize instructions clearly.

## Usage / Flow
1. Define the variables that will change.  
2. Insert them into the template.  
3. Call the LLM — the values are filled in, and the final text is generated automatically.

## Extra Controls
- **Temperature**: controls creativity or determinism in the output.  
  - Low → deterministic / factual output  
  - Medium → neutral  
  - High → more creative / varied output
- Specify style or level of explanation (e.g., "explain as if the user is 5 years old") or language.

---

*Tip:* Using multi-line templates makes complex instructions clearer and easier to maintain.
