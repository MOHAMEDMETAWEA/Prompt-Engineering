# Chat Message Types in LangChain

- **SystemMessage**: Sets the AI's role or behavior (developer-defined).  
- **HumanMessage**: Represents user input/questions.  
- **AIMessage**: Represents the AI's generated response.  

**Usage:** A conversation is a list of messages `[SystemMessage, HumanMessage, AIMessage, ...]` to maintain context for multi-turn chats.


```python
from langchain.chat_models import ChatOpenAI
from langchain.schema import HumanMessage ,AIMessage ,SystemMessage

chat = ChatOpenAI(model_name="gpt-4o-mini", temperature=0.5)


prompt="exlpain langchain"
response = chat([HumanMessage(content=prompt)])
print(response.content)

```

```python
system_context = SystemMessage(
    content="You are a colleague who works in network security and you have to advise me on a problem"
)

human_message = HumanMessage(
    content="I received an email and I think I clicked on a malicious link. How can I be sure that my device has not been hacked?"
)
ai_response=chat([system_context,human_message])
print (ai_response)
```

```python
# System: "You are a chief, you want to ask me some things to find out my favorite Asian dish"
# AI: "Where are you from"
# Human: "Egypt"
# AI: "Do you like spicy food?"
# Human: "Not too much"
# AI: "Do you suffer from diabetes?"
# Human: "yes"

base_conversation = [
    SystemMessage(content="You are a chief, you want to ask me some things to find out my favorite Asian dish"),
    AIMessage(content="Where are you from?"),
    HumanMessage(content="Egypt"),
    AIMessage(content="Do you like spicy food?"),
    HumanMessage(content="Not too much"),
    AIMessage(content="Do you suffer from diabetes?"),
    HumanMessage(content="yes"),
]

ai_response = chat(base_conversation)

print(ai_response)
```

```python
base_convesation = [

    SystemMessage(content="You are a chief, you want to ask me some things to find out my favorite Asian dish, please ask one question")

]

ai_response =None

for i  in range (4):
    ai_response=chat(base_convesation)
    print(ai_response.content)
    base_convesation.append(ai_response)
    user_response=HumanMessage(content=input())

    base_convesation.append(user_response)

human_final_message = HumanMessage(content="And now. Which plate do you think I will like. Just type the plate name")

base_convesation.append(human_final_message)
final_airesponse=chat(base_convesation)
print(final_airesponse)
```
