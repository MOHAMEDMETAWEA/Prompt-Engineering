# Chat

* System: Context or Background information

* Human: Message from user

* AI: Message from llm

# Chat Message Types in LangChain

- **SystemMessage**: Sets the AI's role or behavior (developer-defined).  
- **HumanMessage**: Represents user input/questions.  
- **AIMessage**: Represents the AI's generated response.  

**Usage:** A conversation is a list of messages `[SystemMessage, HumanMessage, AIMessage, ...]` to maintain context for multi-turn chats.

```python
from langchain_core.messages import SystemMessage, AIMessage, HumanMessage
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI

# Create the LLM
llm = ChatOpenAI(
    model="gpt-3.5-turbo",
    temperature=0.7
)

# Define the conversation messages
base_conversation = [
    SystemMessage(content="You are a chef, you want to ask me some things to find out my favorite Asian food."),
    AIMessage(content="Where are you from?"),
    HumanMessage(content="I am from Egypt"),
    AIMessage(content="Do you like spicy food?"),
    HumanMessage(content="Yes, I do"),
    AIMessage(content="Do you suffer from diabetes?"),
    HumanMessage(content="Yes, I do"),
]

# Wrap messages in a ChatPromptTemplate
prompt = ChatPromptTemplate.from_messages(base_conversation)

# Create the chain
chain = prompt | llm

# Invoke the chain
ai_response = chain.invoke({})

# Print AI response
print(ai_response.content)
```

```python
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI

# Initialize the LLM
llm = ChatOpenAI(
    model="gpt-3.5-turbo",
    temperature=0.3  # lower temperature = more accurate & consistent
)

# Enhanced system instruction
system_message = SystemMessage(
    content=(
        "You are a professional Asian chef and nutrition-aware food consultant. "
        "Your goal is to recommend suitable Asian dishes based on the user's preferences, "
        "cultural background, spice tolerance, and health conditions such as diabetes. "
        "Ask concise, relevant questions only if needed, then provide a final recommendation "
        "with brief reasoning."
    )
)

# Conversation history (grounded facts)
conversation = [
    system_message,
    AIMessage(content="Where are you from?"),
    HumanMessage(content="I am from Egypt"),
    AIMessage(content="Do you like spicy food?"),
    HumanMessage(content="Yes, I do"),
    AIMessage(content="Do you suffer from diabetes?"),
    HumanMessage(content="Yes, I do"),
    HumanMessage(
        content=(
            "Based on this information, recommend my favorite Asian food. "
            "Make sure it is suitable for diabetes and explain why."
        )
    )
]

# Build prompt
prompt = ChatPromptTemplate.from_messages(conversation)

# Build chain
chain = prompt | llm

# Run chain
response = chain.invoke({})

# Output result
print(response.content)
```

```python
import os
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage
from langchain_openai import ChatOpenAI

# ==============================
# ðŸ”‘ API CONFIGURATION
# ==============================

# Make sure you set your API key before running:
# setx OPENAI_API_KEY "sk-xxxx"   (Windows)
# export OPENAI_API_KEY="sk-xxxx" (Linux/Mac)

# ==============================
# ðŸ¤– INITIALIZE LLM
# ==============================

llm = ChatOpenAI(
    model="gpt-3.5-turbo",
    temperature=0.7
)

# ==============================
# ðŸ§  INITIAL SYSTEM MESSAGE
# ==============================

conversation = [
    SystemMessage(
        content=(
            "You are a professional Asian chef. "
            "Ask the user relevant questions to discover their favorite Asian food. "
            "Ask ONE clear question at a time."
        )
    )
]

# ==============================
# ðŸ’¬ CHAT LOOP (3 QUESTIONS)
# ==============================

for _ in range(3):
    # AI asks a question (NEW METHOD)
    ai_message = llm.invoke(conversation)
    print(f"AI: {ai_message.content}")

    # Save AI message
    conversation.append(AIMessage(content=ai_message.content))

    # User answers
    user_input = input("You: ")
    conversation.append(HumanMessage(content=user_input))

print("\nâœ… Conversation finished.")
```

```python
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI

# Initialize the LLM
llm = ChatOpenAI(
    model="gpt-3.5-turbo",
    temperature=0.3  # lower temperature = more accurate & consistent
)

# user speak with chat

base_conversation = [ 
    SystemMessage(content="You are a chef, you want to ask me some things to find out my favorite Asian food.")
]

ai_response = None 

for i in range(3):  
    ai_message = llm.invoke(base_conversation)
    print(f"AI: {ai_message.content}")

    # Save AI message
    base_conversation.append(AIMessage(content=ai_message.content))

    # User answers
    user_input = HumanMessage(content=input("You: "))
    base_conversation.append(user_input)

# stop asking and recommend food after 3 questions

human_final_message = HumanMessage(
    content=(
        "Based on this information, recommend my favorite Asian food. "
        "Make sure it is suitable for diabetes and explain why."
    )
)

base_conversation.append(human_final_message)
ai_final_response = llm.invoke(base_conversation)
print(f"AI: {ai_final_response.content}")

print("\nâœ… Conversation finished.")
```

```python
from dotenv import dotenv_values
from langchain_core.messages import SystemMessage, HumanMessage
from langchain_openai import ChatOpenAI

# ==============================
# ðŸ”‘ Load environment variables
# ==============================
env_values = dotenv_values("app.env")

openai_api_key = env_values["OPENAI_API_KEY"]
openai_api_base = env_values.get("OPENAI_API_BASE")

# ==============================
# ðŸ¤– Initialize Chat Model
# ==============================
Chat = ChatOpenAI(
    api_key=openai_api_key,
    base_url=openai_api_base,
    model="gpt-4o-mini",
    temperature=0.7
)

# ==============================
# ðŸ§  Messages
# ==============================
human_message = HumanMessage(
    content=(
        "I received an email and I think I clicked on a malicious link. "
        "What should I do to make sure my device is not hacked?"
    )
)

system_context = SystemMessage(
    content=(
        "You are a cybersecurity expert. Provide clear and concise advice on how to handle potential phishing attacks. "
        "Focus on immediate steps to secure the device and personal information."
    )
)

conversation = [
    system_context,
    human_message
]

# ==============================
# ðŸš€ Invoke the model
# ==============================
response = Chat.invoke(conversation)

print(response.content)
```
